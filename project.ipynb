```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

## Importing all dataset

```python
rt=pd.read_csv("4.1.a ReactionTypes.csv")
reac=pd.read_csv("4.1.b Reactions.csv")
cont=pd.read_csv("4.1.c Content.csv")
```

## <font color='Magenta' face='Cursive'> 1.1 Data Cleaning and Preparation of Reaction Types dataset</font>

```python
rt.info()
#create a copy of data set
rt_cy=rt.copy()
import missingno as msno   # check for missing values visually
msno.matrix(rt_cy)
```

### There are no missing data in ReactonTypes dataset

```python

# Rename the column 'Type' to 'Reaction Type'
rt_cy.rename(columns={'Type': 'Reaction Type'}, inplace=True)
rt_cy.info()
```

## <font color='Magenta' face='Cursive'> 1.2 Data Cleaning and Preparation of Reaction dataset</font>

```python
reac.info()
reac.head(10)
```

```python
#create a copy of data set
reac_cy=reac.copy()
reac_cy.columns
```

```python
import missingno as msno   # check for missing values visually
msno.matrix(reac_cy)
```

```python
reac.info()
reac.head(10)
```

### Droping 'User ID'column as we have no need for it

```python
columns_to_drop = ['User ID']
reac_cy = reac_cy.drop(columns=columns_to_drop)
```

```python
reac_cy.isnull().sum()
```

### There are some null values in "Type column" Droping those null values

```python
reac_cy=reac_cy.dropna(subset=['Type'])
```

```python
reac_cy.isnull().sum()
```

```python

# Rename the column 'Type' to 'Reaction Type'
reac_cy.rename(columns={'Type': 'Reaction Type'}, inplace=True)
reac_cy.info()
```

## <font color='Magenta' face='Cursive'> 1.3Data Cleaning and Preparation of content dataset</font>

```python
cont.info()
cont_cy=cont.copy()
cont_cy.head(10)
```

```python
import missingno as msno   # check for missing values visually
msno.matrix(cont_cy)
```

### Droping 'User ID' and 'URL' column as we have no need of it

```python
columns_to_drop = ['User ID','URL']
cont_cy = cont_cy.drop(columns=columns_to_drop)
```

```python
cont_cy.isnull().sum()
```

```python
# Remove double quotes and parentheses from 'Category' column
cont_cy['Category'] = cont_cy['Category'].apply(lambda x: x.replace('"', '').replace('(', '').replace(')', ''))
```

```python
# Rename the column 'Type' to 'Content Type'
cont_cy.rename(columns={'Type': 'Content Type'}, inplace=True)
cont_cy.info()
```

### <font color = ' Green'>  We have now cleaned all the dataset  </font>

## <font color='orange' face='Cursive'> Now mergeing all three dataset into new dataset called "merged_df"</font>

```python
# Merge the datasets on the 'ID' column
merged_df = pd.merge(cont_cy, reac_cy, on='Content ID')
merged_df = pd.merge(merged_df, rt_cy, on='Reaction Type')

```

```python
columns_to_drop = ['Unnamed: 0_x','Unnamed: 0_y','Unnamed: 0']
merged_df = merged_df.drop(columns=columns_to_drop)
# Convert the 'Category' column to lowercase
merged_df['Category'] = merged_df['Category'].str.lower()
```

```python
merged_df.info()
```

```python
# Save the merged dataset to a new CSV file
merged_df.to_csv('merged_dataset.csv', index=False)
```

```python
import missingno as msno   # check for missing values visually
msno.matrix(merged_df)
```

```python
merged_df.head(10)
```

## <font color='green' face='Cursive'> Data Modeling </font>

```python
# Group by 'Category' and sum the 'Score' for each category
category_scores = merged_df.groupby('Category')['Score'].sum()

# Sort the categories by their total score and select the top 5
top_categories_by_score = category_scores.sort_values(ascending=False).head(5)


# Plot the results
plt.figure(figsize=(10, 6))
ax = top_categories_by_score.plot(kind='bar', color='skyblue')
# Annotate the bars with their values
for container in ax.containers:
    ax.bar_label(container)
plt.title('Top 5 Categories by Total Score')
plt.xlabel('Category')
plt.ylabel('Total Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'Category' and 'Sentiment' and count the occurrences
category_sentiment_counts = merged_df.groupby(['Category', 'Sentiment']).size().unstack(fill_value=0)

# Sum the counts for each category and sort by the total counts to get the top 5 categories
top_categories_by_sentiment = category_sentiment_counts.sum(axis=1).sort_values(ascending=False).head(5)

# Select the rows for the top 5 categories
top_categories_sentiment_data = category_sentiment_counts.loc[top_categories_by_sentiment.index]

# Plot the results with different bars for each sentiment
ax = top_categories_sentiment_data.plot(kind='bar', figsize=(12, 7), color=['skyblue', 'lightgreen', 'salmon'])

# Annotate the bars with their values
for container in ax.containers:
    ax.bar_label(container)

plt.title('Sentiment Distribution in Top 5 Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.show()
```

```python

# Convert 'Datetime' to datetime format
merged_df['Datetime'] = pd.to_datetime(merged_df['Datetime'])

# Filter the dataset to include only entries with positive sentiment
positive_sentiment_df = merged_df[merged_df['Sentiment'] == 'positive']

# Group by 'Category' and sum the scores
category_scores = positive_sentiment_df.groupby('Category')['Score'].sum()

# Sort and get the top 5 categories by score
top_5_categories_by_positive_score = category_scores.sort_values(ascending=False).head(5)

# Plot the results
plt.figure(figsize=(10, 6))
ax = top_5_categories_by_positive_score.plot(kind='bar', color='skyblue')
# Annotate bars with scores
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

plt.title('Top 5 Categories by Positive Sentiment Scores')
plt.xlabel('Category')
plt.ylabel('Total Positive Sentiment Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Get the number of unique categories
unique_categories = merged_df['Category'].unique()
unique_categories_count = len(unique_categories)

print(f'The number of unique categories is: {unique_categories_count}')
print('The unique categories are:')
for category in unique_categories:
    print(category)
```

```python
# Count the number of posts for each unique category
category_counts = merged_df['Category'].value_counts()

# Plot the bar graph
plt.figure(figsize=(20, 8))
ax = category_counts.plot(kind='bar', color='skyblue')
# Annotate bars with scores
for container in ax.containers:
    ax.bar_label(container, label_type='edge')
plt.title('Number of Posts per Category')
plt.xlabel('Category')
plt.ylabel('Number of Posts')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert 'Datetime' to datetime format
merged_df['Datetime'] = pd.to_datetime(merged_df['Datetime'])

# Extract month and year from 'Datetime'
merged_df['YearMonth'] = merged_df['Datetime'].dt.to_period('M')

# Count the number of posts for each month
monthly_post_counts = merged_df['YearMonth'].value_counts().sort_index()

# Identify the month with the most posts
most_posts_month = monthly_post_counts.idxmax()
most_posts_count = monthly_post_counts.max()

# Plot the results
plt.figure(figsize=(12, 6))
ax = monthly_post_counts.plot(kind='bar', color='skyblue')
# Annotate bars with scores
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

# Highlight the month with the most posts
most_posts_index = monthly_post_counts.index.get_loc(most_posts_month)
plt.bar(most_posts_index, most_posts_count, color='red')

plt.title('Number of Posts per Month')
plt.xlabel('Month')
plt.ylabel('Number of Posts')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert 'Datetime' to datetime format
merged_df['Datetime'] = pd.to_datetime(merged_df['Datetime'])

# Group by 'Category' and resample by month, and sum the scores
category_monthly_scores = merged_df.groupby(['Category', pd.Grouper(key='Datetime', freq='M')])['Score'].sum().unstack(fill_value=0)

# Sum the scores for each category and sort by the total scores to get the top 5 categories
top_categories_by_score = category_monthly_scores.sum(axis=1).sort_values(ascending=False).head(5)

# Select the rows for the top 5 categories
top_categories_monthly_data = category_monthly_scores.loc[top_categories_by_score.index]

# Plot the results for monthly aggregation with month names
plt.figure(figsize=(14, 8))
ax = top_categories_monthly_data.T.plot(kind='bar', width=0.8, figsize=(14, 8))
# Annotate bars with scores
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

# Format x-axis to show month names
plt.gca().set_xticklabels([pd.Timestamp(date).strftime('%B %Y') for date in top_categories_monthly_data.columns], rotation=45)

plt.title('Monthly Score Over Time for Top 5 Categories')
plt.xlabel('Month')
plt.ylabel('Score')
plt.legend(title='Category')
plt.show()
```

```python

# Convert 'Datetime' to datetime format
merged_df['Datetime'] = pd.to_datetime(merged_df['Datetime'])

# Filter the data for positive sentiments
positive_sentiments_df = merged_df[merged_df['Sentiment'] == 'positive']

# Group by 'Category' and sum the scores
category_scores = positive_sentiments_df.groupby('Category')['Score'].sum()

# Get the top 5 categories by total score
top_5_categories = category_scores.sort_values(ascending=False).head(5).index

# Filter the data for the top 5 categories
top_5_categories_df = positive_sentiments_df[positive_sentiments_df['Category'].isin(top_5_categories)]

# Group by 'Category' and 'Reaction Type' and count the occurrences
reaction_counts = top_5_categories_df.groupby(['Category', 'Reaction Type']).size().unstack(fill_value=0)

# Plot the bar graph with grouped bars for each reaction type
reaction_counts.plot(kind='bar', width=0.8, figsize=(14, 8))

plt.title('Reaction Types for Positive Sentiment in Top 5 Categories')
plt.xlabel('Category')
plt.ylabel('Count of Reaction Types')
plt.legend(title='Reaction Type')
plt.show()
```
